```python
df = pd.read_csv('file/churn.csv')
df.head()
```

```python
df.Geography.value_counts().plot(kind='bar')
```

```python
df['Exited'].value_counts() / len(df) * 100
```

```python
plt.bar(df['Geography'], df['EstimatedSalary'].mean(), color="cyan", edgecolor="orange")
plt.title("Geography vs EstimatedSalary")
plt.show()
```

```python
plt.bar(df['Gender'], df['Age'].max(), color="cyan", edgecolor="orange")
plt.title("Geography vs EstimatedSalary", fontdict={"fontsize" : 20})
plt.show()
```
